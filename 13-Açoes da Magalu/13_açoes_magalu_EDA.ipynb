# **Criando uma Série Temporal:**

```python
df_serie = df.set_index("Data")
df_serie.head()
```

```python
plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(20,8))
plt.plot(df_serie.index, df_serie["Fechamento"], color="blue", ls="-", label="Fechamento")
plt.plot(df_serie.index, df_serie["Abertura"], color="red", ls=":", label="Abertura")
plt.title("Análise das ações da Magalu - Fechamento | Abertura")
plt.xlabel("Período da cotação")
plt.ylabel("Valor da ação (R$)")
plt.legend();
```

```python
df_serie.tail()
```

```python
media_movel_5Dias = df_serie["Fechamento"].rolling(5).mean()
media_tendencia = df_serie["Fechamento"].rolling(30).mean()
```

```python
plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(20,8))

plt.plot(df_serie.index, df_serie["Fechamento"], color="blue", ls="-", label="Fechamento")
plt.plot(df_serie.index, media_movel_5Dias, color="red", ls=":", label="Média Móvel de 5 dias")
plt.plot(df_serie.index, media_tendencia, color="green", ls="--", label="Tendência")

plt.title("Análise das ações da Magalu - Fechamento | Abertura")
plt.xlabel("Período da cotação")
plt.ylabel("Valor da ação (R$)")
plt.legend();
```

**Boxplot Mensal**

Analisando as oscilações

```python
df["Mes"] = df["Data"].dt.month

plt.figure(figsize=(16, 8))
sns.boxplot(data=df, x="Mes", y="Fechamento");
```

```python
df.groupby(by=["Mes"])["Fechamento"].describe()
```
